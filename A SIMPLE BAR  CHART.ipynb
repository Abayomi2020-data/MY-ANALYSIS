## Importing of the needed library

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

## The datasets used in the chart

```python
company = ["Google", "Amazon", "MSFT", "Facebook"]
revenue = [90,136,89,27]
profit = [40,2,34,12]
```

## Changing of company name to array of numbers  

```python
xpos = np.arange(len(company))
xpos
```

## Making a visualisation of the dataset with bar chart with arrays of numbers and revenue insteads of company name and revenue

```python
plt.title("USA Tech Stocks", fontsize=15)
plt.ylabel("Revenue", fontsize=15)
plt.bar(xpos, revenue)
```

## Making of some adjustment by replacing the arrays of numbers by the company names in the bar chart  

```python
plt.title("USA Tech Stocks", fontsize=15)
plt.xticks(xpos, company)
plt.ylabel("Revenue(bin$)", fontsize=15)
plt.bar(xpos, revenue)
```

## The visualization of of revenue and profit of each company 

```python
plt.title("USA Tech Stocks", fontsize=15)
plt.xticks(xpos, company)
plt.ylabel("Revenue", fontsize=15)
plt.bar(xpos, revenue, label="Revenue")
plt.bar(xpos, profit, label="Profit")
plt.legend(loc="best")
```

```python
plt.title("USA Tech Stocks", fontsize=15)
plt.xticks(xpos, company)
plt.ylabel("Revenue", fontsize=15)
plt.bar(xpos-0.2, revenue,width=0.4, label="Revenue")
plt.bar(xpos+0.2, profit,width=0.4,label="Profit")
plt.legend(loc="best")
```

```python
plt.title("USA Tech Stocks", fontsize=15)
plt.yticks(xpos, company)
plt.ylabel("Revenue", fontsize=15)
plt.barh(xpos-0.2, revenue, label="Revenue")
plt.barh(xpos+0.2, profit, label="Profit")
plt.legend(loc="best")
```
